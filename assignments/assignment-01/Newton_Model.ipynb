```python
import matplotlib.pyplot as plt
```

```python

Ta = 25
steps = 10
k = 0.1
time = 0
t = 0.5
T = 100
lst_time = []
lst_T = []
for i in range(steps):
#     print (T)
    lst_T.append(T)
    lst_time.append(time)
    T = T - k*t*(T-Ta)
    time = time + t


plt.plot(lst_time, lst_T, 'ro')
plt.axis([0, time, 0, 100])
plt.xlabel('time')
plt.ylabel('Temperature')  
plt.title('Question 1 part d')
plt.show()
    
```

```python
Ta = 25
steps = 10
k = 0.4
time = 0
t = 0.5
T = 100
e = 2.718
C = 75
lst_time = []
lst_T = []
lst_actual = []
for i in range(steps):
#     print (T)
    lst_actual.append((e**(-k*time - C)) + Ta)
    lst_T.append(T)
    lst_time.append(time)
    T = T - k*t*(T-Ta)
    time = time + t

# plt.plot(time, time, 'r--', time, time**2, 'bs', time, time**3, 'g^')
plt.plot(lst_time, lst_T, 'ro', lst_time, lst_actual, "r--")
plt.axis([0, time, 0, 100])
plt.xlabel('time')
plt.ylabel('Temperature')  
plt.title('Question 1 part e')
plt.show()
```
