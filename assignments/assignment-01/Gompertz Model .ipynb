```python
import matplotlib.pyplot as plt
import math
```

```python

e = 2.718281
M = 1000000
steps = 100
k = 100
time = 0
t = 0.001
N = 1000
lst_time = []
lst_N = []
for i in range(steps):
#     print (T)
    lst_N.append(N)
    lst_time.append(time)
#     print(N)
    N = N + k*t*N* (math.log(e,int(M / N)))
    time = time + t


plt.plot(lst_time, lst_N, 'r--')
plt.axis([0, time, 0, N])
plt.xlabel('time')
plt.ylabel('Tumor')  
plt.title('Question 2 part b')
plt.show()
```
